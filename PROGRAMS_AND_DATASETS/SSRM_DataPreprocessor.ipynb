```python
#import the csv file
#splite to rows
#remove the 1st row
#`split the 2nd row based on the comma
# added to list all unique symtoms to the list.
# create new csv with this many

import csv
import pandas as pd

lst = list()
dflst = list()
with open('DerivedKnowledgeGraph_final.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for row in csv_reader:
        row_dit={}
        row_dit['disease']=row[0]
        for element in row[1].split(','):
            ele=element.split('(')[0].strip()
            val = element.split('(')[1].split(')')[0].strip()

            if ele not in lst:
                lst.append(ele)
            #values
            row_dit[ele]=val

        dflst+=[row_dit]
        #row
lst=['Disease']+lst
```

```python
df = pd.DataFrame(dflst)
df.set_index('disease')
```

```python
df.to_csv('new_csv.csv', sep=',', encoding='utf-8',index=False)
```
