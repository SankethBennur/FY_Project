```python
import pandas as pd
import numpy as np
```

```python
data =  pd.read_csv('sample_testing.csv', delimiter = ',')
```

```python
data = data.replace(np.nan,'0')
```

```python
datat = pd.get_dummies(data.disease)
```

```python
#print(datat[data.disease])
lst = {}
for d in data.disease:
    st = ''
    for c in datat[d]:
        st+=str(c)
    print(st,d)
    lst[d]=st
```

```python
#data.disease = lst[data.disease]
data = data.replace({"disease": lst})
```

```python
sample = pd.DataFrame(data)
```

```python
sample
```

```python
sample = sample.set_index(sample.disease)
sample = sample.drop(columns=['disease'])
```

```python
for ro in sample:
    for e in sample[ro].index.values:
        if str(sample[ro][e])>'0':
            sample[ro][e]=1
```

```python
sample.to_csv('NDS.csv', sep=',')
```
